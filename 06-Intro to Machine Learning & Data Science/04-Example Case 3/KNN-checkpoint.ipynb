```python
from sklearn.cross_validation import train_test_splitst_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=123)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_1 = KNeighborsClassifier(n_neighbors = 1)
knn_1.fit(x_train, y_train)
```

```python
knn_1.score(x_train, y_train)
```

```python
knn_1.score(x_test, y_test)
```

```python
knn_2 = KNeighborsClassifier(n_neighbors = 2)
knn_2.fit(x_train, y_train)
```

```python
knn_1.score(x_test, y_test)
```

```python
params = {'n_neighbors': [1, 3, 4, 5, 15, 25, 50, 100, 200, 500]}

from sklearn.grid_search import GridSearchCV
knn = KNeighborsClassifier(weights='distance')
gsCV = GridSearchCV(knn, param_grid = params, cv = 3)
gsCV.fit(x_train, y_train)
```

```python
gsCV.best_params_
```

```python
gsCV.best_score_
```

```python
gsCV.score(x_test, y_test)
```
